```python
df = pd.read_csv("http://stats.moe.gov.tw/files/detail/107/107_student.csv")
```

```python
df.head()
```

```python
colnames = df.columns
col = np.array(colnames)
colnames,col
```

```python
(df['日間∕進修別'] == 'D 日') & (df['等級別'] == 'B 學士')
```

```python
df1 = df[(df['日間∕進修別'] == 'D 日') & (df['等級別'] == 'B 學士')]
```

```python
df1.head()
```

```python
df_girl = df1.loc[:,['學校名稱','一年級女生','二年級女生','三年級女生','四年級女生']]
df_girl
```

```python
df_boy = df1.loc[:,['學校名稱','一年級男生','二年級男生','三年級男生','四年級男生']]
df_boy
```

```python
df_girl.sum(axis = 0)  # data type is not integer!
```

```python
df_girl.columns = ['學校名稱', '一年級', '二年級', '三年級', '四年級']
df_girl.head()
```

```python
df_boy.columns = ['學校名稱', '一年級', '二年級', '三年級', '四年級']
df_boy.head()
```

```python
df_girl_sum = df_girl[['一年級', '二年級', '三年級', '四年級']].astype(int,errors = 'ignore').sum()
df_girl_sum
```

```python
df_boy_sum = df_boy[['一年級', '二年級', '三年級', '四年級']].astype(int,errors = 'ignore').sum()
df_boy_sum
```

```python
plt.plot(df_boy_sum)
plt.plot(df_girl_sum)
plt.ylabel('人數')
plt.legend(['男生人數','女生人數'], loc = 'best')
```
