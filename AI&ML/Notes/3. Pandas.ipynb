### 解決 TypeError: unsupported operand type(s) for /: 'str' and 'str'
### 將str變成int

```python
import locale

# atof
```

```python
locale.setlocale(locale.LC_NUMERIC,'')
```

```python
df2['一年級男生'] = df2['一年級男生'].apply(locale.atof)
```

```python
df2['一年級女生'] = df2['一年級女生'].apply(locale.atof)
```

```python
df2.一年級男生/df2.一年級女生
```

```python
df2['一年級男女比'] = df2.一年級男生/df2.一年級女生
```

```python
df2.head()
```

```python
df2.sort_values(by = ['一年級男女比'],ascending = False) #ascending 上升
```

```python
df3 = df2.sort_values(by = ['一年級男女比'],ascending = False)
```

```python
df3[:20].plot.bar(x = '學校名稱', y = ["一年級男女比"]) 
```
