# L2助教課

```python
x = [1,2,3,4]
y1 = [5,9,7,9]
y2 = [5,7,8,8]
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
plt.figure(figsize =(5,8) )
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
plt.figure(figsize =(5,3) , facecolor = "b")
plt.plot(x,y1)
plt.plot(x,y2)
```

```python
plt.figure(figsize =(5,3))
plt.plot(x, y1, label = 'y1')
plt.plot(x, y2, label = 'y2')

#plt.legend(loc = "upper left")  #Location of the label 
plt.legend(loc = "best")
```

```python
plt.figure(figsize =(5,3))
plt.plot(x, y1, label = 'y1')
plt.plot(x, y2, label = 'y2')

plt.legend(loc = "best")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.title('Title')
```

```python
plt.figure(figsize =(5,3))
plt.plot(x, y1, label = 'y1')
plt.plot(x, y2, label = 'y2')

plt.legend(loc = "best")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

plt.xlim(-5,5) #軸的距離
plt.ylim(-10,10)

plt.xticks(np.arange(-5,5,2))# 距離,間距
plt.yticks(np.arange(-10,10,3))
plt.title('Title')
```

# Seaborn

```python
import seaborn as sns
```

```python
sns.set_style('whitegrid') #白底
plt.plot(x, y1)
plt.plot(x, y2)
```

```python
tips = sns.load_dataset('tips')
tips.head() #call the data set
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = tips, kind = 'scatter')
#(x,y為固定名稱的參數, data = 資料庫, kind = 圖形種類，此為散布圖)
```

```python
sns.pairplot(tips)
```
