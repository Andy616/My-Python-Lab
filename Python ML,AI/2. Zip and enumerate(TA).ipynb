```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## 畫一張圖

```python
x = np.linspace(0,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
#設定圖形大小是(6，6)
plt.figure(figsize = (6,6), facecolor = "r") 
#設定一個風格
plt.xkcd()
#要先設定再畫圖，否則會跑掉!!


#設定圖例(放在左下角)
plt.plot(x,y1,label = "sin(x)")
plt.plot(x,y2,label = "cos(x)")
plt.legend(loc = "lower left")

#設定X，Y軸範圍
plt.xlim(0,10)
plt.ylim(-1,1)

#設定X，Y軸名稱
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

#X軸分成10個部分，Y軸分成4個部分
plt.xticks(np.arange(0,10,1)) #第三個參數為間距
plt.yticks(np.arange(-1,1,0.5))

#圖的標題是"Sin and Cos"
plt.title("Sin and Cos")
```

## Variable Length Function Arguments

```python
def f(x):
    print(x)
```

```python
f('Hello')
```

```python
f('Hello','World')
```

```python
def ARGV(*argv):  #加上*輸入不受限制 , argv is a list
    for i in argv:
        print(i)
```

```python
ARGV('Hello','World')
```

```python
def ARGV1(arg1, *argv):
    print(arg1)
    print(" ")
    for i in argv:
        print(i)
```

```python
ARGV1("a","b","c","d",'e') #以空格表示不同系統
```

```python
def ARGV1(*argv , arg1):
    print(arg1)
    print(" ")
    for i in argv:
        print(i)
```

```python
ARGV1("a","b","c","d",'e')
```

```python
ARGV1("a","b","c","d",arg1='e')
```

## Variable Length Keywoed Argument

```python
def KWARG(**kwargs):
    for key,value in kwargs.items():  # .item 從字典中取出東西
        print(f'{key} == {value}')
```

```python
KWARG(first = 'a',second = 'b',third = 'c')
```

```python
def Three(arg1,arg2,arg3):
    print('arg1 is ',arg1)
    print('arg2 is ',arg2)
    print('arg3 is ',arg3)
```

```python
args = ('a','b','c')
```

```python
Three(*args)
```

```python
kwargs = {'arg1':'a','arg3':'c','arg2':'b'}
```

```python
Three(**kwargs)
```

# map and lambda

```python
def square(x):
    return x**2
```

```python
list(map(square,[1,2,3,4,5]))
```

```python
list(map(lambda x: x**2, [1,2,3,4,5]))  #lambda 有點像內建函式
```
